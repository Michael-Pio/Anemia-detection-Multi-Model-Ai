Generate Artificial Data

```python
import csv
import random

# Define the symptoms and their respective probabilities
symptoms = [
    "General_Fatigue",
    "Weakness",
    "Shortness_of_Breath",
    "Pale_Skin",
    "Rapid_Heartbeat",
    "Dizziness_or_Lightheadedness",
    "Cold_Hands_and_Feet",
    "Brittle_Nails",
    "Headaches",
    "Poor_Concentration",
    "Chest_Pain",
    "Restless_Legs"
]

# Define the base probabilities for each symptom
base_probabilities = [0.5, 0.45, 0.3, 0.1, 0.25, 0.2, 0.15, 0.05, 0.4, 0.35, 0.15, 0.1]

# Generate data for each participant
total_participants = 526  # Total number of participants (100 with anemia, 100 without anemia)
participants_with_anemia = 322

# Calculate the number of participants without anemia
participants_without_anemia = total_participants - participants_with_anemia

# Generate data for participants with anemia
data_anemia = []
for _ in range(participants_with_anemia):
    participant_data = [1 if random.random() < prob else 0 for prob in base_probabilities]
    data_anemia.append([1] + participant_data)  # Add a label "1" for participants with anemia

# Generate data for participants without anemia
data_without_anemia = []
for _ in range(participants_without_anemia):
    participant_data = [1 if random.random() < (prob-0.2) else 0 for prob in base_probabilities ] 
    data_without_anemia.append([0] + participant_data)  # Add a label "0" for participants without anemia

# Combine the two datasets
data = data_anemia + data_without_anemia

# Create the CSV file
csv_file_path = "anemia_symptom_data_with_labels.csv"
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Write the header row
    writer.writerow(['Label'] + symptoms)

    # Write the data rows
    writer.writerows(data)

print(f"CSV file '{csv_file_path}' has been created with nearly accurate data and labels.")
```

```python
import os
FilesList = []
Original = []
Anemic= []
NonAnemic = []
FilesList = os.listdir("DataSet\Palm")

for key,ele in enumerate(FilesList):
    if(ele.find("(")==-1):
        Original.append(ele)
        if(ele.find("Non-Anemic")==-1):
            Anemic.append(ele)
        else:
            NonAnemic.append(ele)

Original
Duplicate = list(set(FilesList) - set(Original))
print(*[O[0:-4]+"\n" for O in Duplicate] )
print(len(Original),len(Anemic),len(NonAnemic))
```

Creating dictionary for Data Augmentation

```python
Dups_Grouped = dict()


for key,ele in enumerate(Original):
    group = []

    for j,dup in enumerate(Duplicate):
        if(dup[0:-4].find(ele[0:-4]) != -1):
            group.append(dup)
    Dups_Grouped[ele]=group

print(Dups_Grouped)
```

```python
import csv
Data_To_Update = []

temp_file_Content = []

first = True
def getList(fileName:str,key:str):
    for row in temp_file_Content:
        if(row['Palm'][0:-4].find(key[0:-4]) != -1):
            val = row
            val['Palm'] = fileName
            return val
    

dupFile = open("anemia_Augmented.csv",mode = "w",newline='')

with open("anemia_symptom_data_with_labels_Palm.csv", 'r', newline='') as temp_file:
    reader = csv.DictReader(temp_file)
    writer = csv.DictWriter(dupFile,fieldnames=reader.fieldnames)
    for row in reader:
        for value in Dups_Grouped[row['Palm']]:
            
            res = row
            res['Palm'] = value
            writer.writerow(res)
            Data_To_Update.append(res)
    
# print(temp_file_Content)
# for key in Dups_Grouped.keys():
#     for file in Dups_Grouped[key] :
#         Data_To_Update.append(getList(file,key))
        
Data_To_Update        
```
